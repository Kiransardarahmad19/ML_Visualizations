```python
import pandas as pd
import numpy as np
file_path = "supermarket_sales.csv"
df = pd.read_csv(file_path, encoding='latin-1')
df.head()


import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.violinplot(x='City', y='Quantity', data=df, palette='pastel')
plt.title('Violin Plot of Quantity by City')
plt.xlabel('City')
plt.ylabel('Quantity')
plt.grid(True)
plt.show()
```
